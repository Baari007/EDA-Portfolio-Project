```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("aerofit_treadmill.csv")

# Display first few rows
df.head()

```

```python
# Check the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Check data types of each column
df.info()

# Check for missing values
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
# Check for duplicate rows
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

```python
# Display first few rows of the dataset again to review data
df.head()
```

```python
# Statistical summary for numerical features
print("Statistical Summary for Numerical Features:")
print(df.describe())

# Statistical summary for categorical features
print("\nStatistical Summary for Categorical Features:")
print(df.describe(include=['object']))
```

```python
# Value counts for all categorical features
print("Value Counts for Product:\n", df['Product'].value_counts())
print("\nValue Counts for Gender:\n", df['Gender'].value_counts())
print("\nValue Counts for MaritalStatus:\n", df['MaritalStatus'].value_counts())
```

```python
# Unique values for all categorical features
print("Unique values in Product:", df['Product'].unique())
print("Unique values in Gender:", df['Gender'].unique())
print("Unique values in MaritalStatus:", df['MaritalStatus'].unique())
```

```python
# Distribution plot for numerical features
plt.figure(figsize=(12, 8))

# Distribution plot for Age
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Distribution of Age')

# Distribution plot for Income
plt.subplot(2, 2, 2)
sns.histplot(df['Income'], kde=True, color='green')
plt.title('Distribution of Income')

# Distribution plot for Usage
plt.subplot(2, 2, 3)
sns.histplot(df['Usage'], kde=True, color='red')
plt.title('Distribution of Usage')

# Distribution plot for Miles
plt.subplot(2, 2, 4)
sns.histplot(df['Miles'], kde=True, color='purple')
plt.title('Distribution of Miles')

plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(12, 8))

# Box plot for Age
plt.subplot(2, 2, 1)
sns.boxplot(x=df['Age'], color='blue')
plt.title('Box plot of Age')

# Box plot for Income
plt.subplot(2, 2, 2)
sns.boxplot(x=df['Income'], color='green')
plt.title('Box plot of Income')

# Box plot for Usage
plt.subplot(2, 2, 3)
sns.boxplot(x=df['Usage'], color='red')
plt.title('Box plot of Usage')

# Box plot for Miles
plt.subplot(2, 2, 4)
sns.boxplot(x=df['Miles'], color='purple')
plt.title('Box plot of Miles')

plt.tight_layout()
plt.show()
```

```python
# Count plot for Product
plt.figure(figsize=(6, 4))
sns.countplot(x='Product', data=df, palette='viridis')
plt.title('Count of Products Purchased')
plt.show()

# Count plot for Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Count of Customers by Gender')
plt.show()

# Count plot for Marital Status
plt.figure(figsize=(6, 4))
sns.countplot(x='MaritalStatus', data=df, palette='viridis')
plt.title('Count of Customers by Marital Status')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate correlation matrix for numeric columns
corr_matrix = numeric_df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numerical column
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Display the outliers
print("Outliers Detected in the Dataset:")
print(outliers.sum())
```

```python
# 7.1 Product – Gender
# Percentage of Male customers purchasing a treadmill
gender_product = pd.crosstab(df['Product'], df['Gender'], normalize='index') * 100
male_percentage = gender_product.loc[:, 'Male'].sum()

# Percentage of Female customers purchasing KP781
female_kp781_percentage = gender_product.loc['KP781', 'Female']

# Probability of a customer being Female given that Product is KP281
prob_female_given_kp281 = gender_product.loc['KP281', 'Female']

print(f"Percentage of Male customers purchasing a treadmill: {male_percentage:.2f}%")
print(f"Percentage of Female customers purchasing KP781: {female_kp781_percentage:.2f}%")
print(f"Probability of a customer being Female given that Product is KP281: {prob_female_given_kp281:.2f}")
```

```python
# 7.2 Product – Age
# Percentage of customers with Age between 20s and 30s among all customers
age_range_percentage = (df[(df['Age'] >= 20) & (df['Age'] <= 30)]['Product'].count() / df['Product'].count()) * 100
print(f"Percentage of customers with Age between 20s and 30s: {age_range_percentage:.2f}%")
```

```python
# 7.3 Product – Income
# Percentage of low-income customers purchasing a treadmill
low_income_percentage = (df[df['Income'] < 30000]['Product'].count() / df['Product'].count()) * 100

# Percentage of high-income customers purchasing KP781
high_income_kp781_percentage = (df[(df['Income'] >= 100000) & (df['Product'] == 'KP781')].shape[0] / df[df['Income'] >= 100000].shape[0]) * 100

# Percentage of high-income customers purchasing KP781 given that Product is KP781
prob_high_income_given_kp781 = (df[(df['Income'] >= 100000) & (df['Product'] == 'KP781')].shape[0] / df[df['Product'] == 'KP781'].shape[0]) * 100

print(f"Percentage of low-income customers purchasing a treadmill: {low_income_percentage:.2f}%")
print(f"Percentage of high-income customers purchasing KP781: {high_income_kp781_percentage:.2f}%")
print(f"Percentage of high-income customers purchasing KP781 given that Product is KP781: {prob_high_income_given_kp781:.2f}")
```

```python
# 7.4 Product – Fitness
# Percentage of customers with Fitness Level 5
fitness_level_5_percentage = (df[df['Fitness'] == 5]['Product'].count() / df['Product'].count()) * 100

# Percentage of customers with Fitness Level 5 purchasing KP781
fitness_level_5_kp781_percentage = (df[(df['Fitness'] == 5) & (df['Product'] == 'KP781')].shape[0] / df[df['Fitness'] == 5].shape[0]) * 100

# Percentage of Fitness Level 5 customers buying KP781
fitness_level_5_buy_kp781_percentage = (df[(df['Fitness'] == 5) & (df['Product'] == 'KP781')].shape[0] / df.shape[0]) * 100

print(f"Percentage of customers with Fitness Level 5: {fitness_level_5_percentage:.2f}%")
print(f"Percentage of customers with Fitness Level 5 purchasing KP781: {fitness_level_5_kp781_percentage:.2f}%")
print(f"Percentage of Fitness Level 5 customers buying KP781: {fitness_level_5_buy_kp781_percentage:.2f}%")
```

```python
# 7.5 Product - Marital Status
# Percentage of customers who are partnered using treadmills
partnered_customers_percentage = (df[df['MaritalStatus'] == 'partnered']['Product'].count() / df['Product'].count()) * 100
print(f"Percentage of customers who are partnered using treadmills: {partnered_customers_percentage:.2f}%")
```
